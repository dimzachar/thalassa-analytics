# Thalassa Exploration

Starter notebook for exploring the local DuckDB tables created by the Bruin pipeline.

```python
from pathlib import Path

import duckdb
import pandas as pd
```

```python
PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
DB_PATH = PROJECT_ROOT / 'duckdb.db'

con = duckdb.connect(str(DB_PATH), read_only=True)
DB_PATH
```

```python
con.sql("show tables").df()
```

```python
daily_counts = con.sql(
    '''
    select *
    from reports.row_counts_daily
    order by service_date
    '''
).df()

daily_counts
```

```python
top_routes = con.sql(
    '''
    select *
    from marts.fct_route_traffic_daily
    order by total_passengers desc
    limit 20
    '''
).df()
top_routes
```

```python
top_ports = con.sql(
    '''
    select *
    from marts.fct_port_activity_daily
    order by departures_count desc, arrivals_count desc
    limit 20
    '''
).df()
top_ports
```
